```python
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
import os
```

```python
# Read the cleaned CSV files and load the tables
# Main Tables

dim_fund = pd.read_csv("../data/processed/clean_fund_master.csv")
fact_nav = pd.read_csv("../data/processed/clean_nav.csv")
fact_transactions = pd.read_csv("../data/processed/clean_transactions.csv")
fact_performance = pd.read_csv("../data/processed/clean_performance.csv")
```

```python
# Remaining cleaned datasets
import pandas as pd

fund_house = pd.read_csv("../data/processed/clean_aum_by_fund_house.csv")
monthly_sip = pd.read_csv("../data/processed/clean_monthly_sip_inflows.csv")
category = pd.read_csv("../data/processed/clean_category_inflows.csv")
industry = pd.read_csv("../data/processed/clean_industry_folio_count.csv")
holdings = pd.read_csv("../data/processed/clean_portfolio_holdings.csv")
benchmark = pd.read_csv("../data/processed/clean_benchmark_indices.csv")
```

```python
# Create dim_date

dates = pd.to_datetime(fact_nav["date"])

dim_date = pd.DataFrame({
    "date": dates.dt.date,
    "year": dates.dt.year,
    "quarter": dates.dt.quarter,
    "month": dates.dt.month,
    "month_name": dates.dt.month_name()
})

dim_date = dim_date.drop_duplicates()
print(dim_date.head())
print(dim_date.shape)
```

```python
#  Load the data into tables

dim_fund.to_sql("dim_fund", engine, if_exists="append", index=False)
```

```python
#  Load the  fact_nav data into tables

fact_nav.to_sql("fact_nav", engine, if_exists="append", index=False)
```

```python
#  Load the fact_transactions data into tables

fact_transactions.to_sql("fact_transactions", engine, if_exists="append", index=False)
```

```python
#  Load the fact_performance data into tables

fact_performance.to_sql("fact_performance", engine, if_exists="append", index=False)
```
